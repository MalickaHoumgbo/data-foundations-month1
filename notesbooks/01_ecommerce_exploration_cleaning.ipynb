```python
# diagnostic des valeurs manquantes
""" on procède à cette étape pour examiner l'entendu des problèmes
Avant de procéder à un état avant/après pour voir ce qui a été supprimé ou comblé"""

import pandas as pd
sales = pd.read_csv("coffee_sales.csv")

# nombre des valeurs manquantes par colonnes:
print(sales.isnull().sum())
print("\n")

# pourcentage des valeurs manquantes
print(sales.isnull().sum()/len(sales)*100)    # on mesure le pourcentage pour estimer à quel point les NAN peuvent être impactants
print("\n")                                   # 89 NAN sur une taille de 100 veux dire 89% de la colonne vide , donc inutilisable
                                              # 89 NAN sur une taille de 3636 veut dire 2.44% de valeurs nulles sur une colonne, c'est encore exploitable

# nombre de valeurs manquantes dans le dataset
print("nombre total de NaN:",sales.isnull().sum().sum(),"\n")  # on vérifie si le dataset est globalement sain ou pas

# vérification des valeurs d'ensemble avant le nettoyage
print("avant nettoyage", sales.shape)
```

```python
## Bilan diagnostic des NAN

### Coffee Sales
"""
- 89 NaN sur la colonne `card` soit 2.4% du dataset
- Cause identifiee: transactions payees en especes (cash)
- Décision : fillna()  remplacer par 'cash' ou 'inconnu', aucune ligne à perdre"""

### Conclusion 
"""
Le diagnostic NaN n'appelle pas toujours une action.
Sur 3 datasets, un seul comportait un cas réel à traiter.
Évaluer l'impact avant d'agir , le volume seul ne suffit pas,
c'est le sens de la donnée qui détermine la décision."
```

```python
# Remplacement des valeurs manquantes avec fillna()

import pandas as pd
sales = pd.read_csv("coffee_sales.csv")

# remplacer les NaN de la colonne card par "Inconnu"
sales["card"] = sales["card"].fillna("inconnu")

# vérifier qu'il ne reste plus de Nan
print(sales.isnull().sum())
print("\n")
print("après fillna", sales.shape)

#Ce que j'ai fait 
"""J'ai remplacé les 89 NaN de la colonne card par la valeur 'inconnu' avec fillna()"""

#Pourquoi?
"""Les NaN correspondent à des paiements en espèces, ce sont des transactions valides. 
Supprimer ces lignes aurait fait perdre de la donnée réelle sans raison"""

#Que m'a appris ça ? 
"""fillna() conserve toutes les lignes contrairement à dropna(). 
Le choix entre les deux dépend du sens de la donnée, pas du volume de NaN"""
```

```python
# corriger les types de colonnes

import pandas as pd
sales = pd.read_csv("coffee_sales.csv")

# inspecter les types de données actuelles
print(sales.dtypes,"\n")

# on remarque que les colonnes date et datetime ont été classées comme texte par pandas
sales['date'] = pd.to_datetime(sales['date'])
sales['datetime'] = pd.to_datetime(sales['datetime'])
print(sales.dtypes,"\n")

# Extraire des informations temporelles
sales['annee'] = sales['date'].dt.year
sales['mois'] = sales['date'].dt.month
sales['jour_semaine'] = sales['date'].dt.day_name()

print("\n")
print(sales[['date','annee','mois','jour_semaine']].head())
```

```python
## corrections des types de colonnes

#Ce que j'ai fait :
"""Converti les colonnes date et datetime de object vers datetime64[ns]"""

#Pourquoi :
"""Pandas charge les dates comme texte par défaut.
Sans conversion, on ne peut ni trier chronologiquement
ni extraire l'année, le mois, le jour."

#Ce que ça m'a appris :
"""to_datetime() est indispensable sur toute colonne de dates.
Après conversion on peut extraire année, mois, jour
avec .dt.year / .dt.month / .dt.day_name()"""
```
