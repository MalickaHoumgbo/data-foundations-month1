```python
# corriger les types de colonnes

import pandas as pd
voyageurs = pd.read_csv("comptage-voyageurs-trains-transilien.csv",
            sep=";",
            encoding = "utf-8-sig")


# Vérifier les types actuels 
print(voyageurs.dtypes) 
print("\n")
# on remarque que la colonne date a été classée comme texte par pandas, il faut rectifier le bon format avec to_datetime

voyageurs['Date'] = pd.to_datetime(voyageurs['Date'])

print(voyageurs.dtypes,"\n")


# Extraire des informations temporelles

voyageurs['mois'] = voyageurs['Date'].dt.month
voyageurs['jour_semaine'] = voyageurs['Date'].dt.day_name()

print("\n")
print(voyageurs[['Date','Annee','mois','jour_semaine']].head())
```
