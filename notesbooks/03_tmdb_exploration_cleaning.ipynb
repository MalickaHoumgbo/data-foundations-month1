```python
## Bilan diagnostic des NAN

### TMDB Movies
- 2 NaN sur une colonne inutile (poster_path)
- La colonne sera supprimee avec drop(), le probleme disparait de lui-meme
```

```python
# utilisation de drop pour supprimer des colonnes inutiles

import pandas as pd
films = pd.read_csv("TMDB_Data.csv")

films = films.drop(["poster_path", "Unnamed: 0"], axis=1)  # Les 2 NaN ont disparu avec la colonne poster_path

# vérifier qu'il ne reste que des données utiles
print(films.isnull().sum())
print("\n")
print("après drop: ", films.shape)
```

```python
## Traitement des valeurs manquantes :TMDB Movies
"""
Diagnostic : 2 NaN uniquement sur la colonne `poster_path`

Décision : ni dropna() ni fillna()  mais drop()

Justification :
- `poster_path` est un chemin d'image sans valeur analytique
- Supprimer la colonne entière règle les 2 NaN automatiquement

- `Unnamed: 0` supprimée au même passage — index doublon inutile
- 0 ligne perdue, le dataset passe de 11 à 9 colonnes

Ce que ça m'a appris :
Quand les NaN se trouvent sur une colonne inutile,
drop() est plus efficace que dropna() ou fillna().
On règle deux problèmes en une seule commande.
La décision dépend toujours du sens de la donnée,
pas uniquement de la présence de NaN. """
```

```python
# détecter et supprimer les doublons 

import pandas as pd
films = pd.read_csv("TMDB_Data.csv")

films = films.drop(["poster_path", "Unnamed: 0"], axis=1) # on supprime les colonnes inutiles
print("\n")

# compter les doublons dans le dataset 
print("doublons totaux: ", films.duplicated().sum()) #Elle compte le nombre de lignes entièrement identiques — 
                                                     #C'est-à-dire des lignes où TOUTES les colonnes ont la même valeur. 
                                                     #Pas colonne par colonne, mais ligne entière contre ligne entière.
print("\n")

# doublons sur une colonne spécifique
print("doublons sur Title", films.duplicated('title').sum())

# afficher les titres dupliqués pour les inspecter
films[films.duplicated('title', keep=False)].head(10)

print("doublons sur id:", films.duplicated('id').sum(),"\n")  # pour voir combien d'ids reviennent plusieurs fois


# Afficher les lignes avec des id dupliqués
films[films.duplicated('id', keep=False)].sort_values('id').head(20)
print("\n")

# Vérifier si ce sont vraiment les mêmes films
print(films[films.duplicated('id', keep=False)][['id', 'title', 'release_date']].sort_values('id').head(20))

# supprimer des doublons avec les memes id car c'est bien des doublons identiques
films_clean = films.drop_duplicates(subset=['id']) 
print("\n")


# pour vérifier combien de ligne ont été supprimés
print("Avant :", films.shape)
print("Après :", films_clean.shape)
print("Lignes supprimées :", len(films) - len(films_clean))
```

```python
## Traitement des doublons — TMDB Movies

#Ce que j'ai fait :
"""Supprimé 8820 lignes dupliquées avec drop_duplicates(subset=['id'])
"""

#Pourquoi :
"""Chaque film apparaissait exactement deux fois dans le dataset
avec le même id, même titre, même date — vrais doublons confirmés
par inspection visuelle avant suppression
"""

#Ce que ça m'a appris :
"""
1. Toujours supprimer les colonnes d'index inutiles AVANT
   de chercher les doublons — Unnamed: 0 masquait les vrais
   Doublons et faussait le diagnostic (duplicated() affichait 0)

2. duplicated().sum() détecte les lignes entièrement identiques
   duplicated('col') détecte les valeurs répétées sur une colonne
   Les deux ensemble donnent un diagnostic complet

3. keep=False dans l'inspection pour voir TOUTES les occurrences
   keep='first' dans drop_duplicates() pour garder l'original
"""
```

```python
# inspecter et corriger les types de colonnes

import pandas as pd
films = pd.read_csv("TMDB_Data.csv")

# inspecter les types de données actuelles
print(films.dtypes,"\n")

# on remarque que les colonnes date et datetime ont été classées comme texte par pandas
films['release_date'] = pd.to_datetime(films['release_date'])
print(films.dtypes,"\n")

# Extraire des informations temporelles
films['annee'] = films['release_date'].dt.year
films['mois'] = films['release_date'].dt.month
films['jour_semaine'] = films['release_date'].dt.day_name()

print("\n")
print(films[['release_date','annee','mois','jour_semaine']].head())
```

```python
## bilan semaine 2

import pandas as pd
films = pd.read_csv("TMDB_Data.csv")

films = films.drop(["poster_path", "Unnamed: 0"], axis=1)
films = films.drop_duplicates(subset=['id'])
films['release_date'] = pd.to_datetime(films['release_date'])
films['annee'] = films['release_date'].dt.year
films['mois'] = films['release_date'].dt.month
films['jour_semaine'] = films['release_date'].dt.day_name()

films.to_csv("tmdb_clean.csv", index=False)
print("Export OK :", films.shape)
```
